Loading the Players Raw Data Set

```python
import pandas as pd
import numpy as np

print("Loading raw data files...")
players_df = pd.read_csv("afl_players_info_raw.csv")
stats_df = pd.read_csv("afl_players_seasonal_stats_raw.csv")
print("Data sets loaded successfully.")
print("=" * 80)
print("Data set Dimensions:")
print(f"Players Info: {players_df.shape}")
print(f"Players Stats: {stats_df.shape}")
print("=" * 80)
```

Inspecting the data

```python
def display_dataframe_info(df, df_name):
    print("=" * 80)
    print(f"{df_name}:")
    print("=" * 80)
    print(df.info())

display_dataframe_info(players_df, "Players Info")
display_dataframe_info(stats_df, "Players Stats")
```

Statisticals

```python
def statistics_summary(df, df_name):
    print("=" * 80)
    print(f"{df_name} numerical statistics")
    print("=" * 80)
    print(df.describe())

statistics_summary(players_df, "Players")
statistics_summary(stats_df, "Players Stats")
```

Finding Missing Values

```python
def access_data_missing_values(df, dataset_name):
    print("=" * 80)
    print(f"Missing Data for: {dataset_name}")
    print("=" * 80)
    missing_data = df.isnull().sum()
    missing_percent = (missing_data / len(df)) * 100
    missing_report = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_percent})
    print(missing_report[missing_report['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False))

    if missing_report['Missing Values'].sum() == 0:
        print("No missing values found.")

access_data_missing_values(players_df, "Players Info DataFrame")
access_data_missing_values(stats_df, "Player Stats DataFrame")
```

Finding Duplicates

```python
def check_duplicates(df, df_name):
    print("2. DUPLICATE RECORDS")
    print("-" * 80)
    duplicates = df.duplicated().sum()
    print(f"Number of duplicate records in {df_name}: {duplicates}\n")
    

check_duplicates(players_df, "Players Info")
check_duplicates(stats_df, "Players Stats")
```

Removing Duplicates

```python
def remove_duplicates(df, df_name):
    df = df.drop_duplicates(inplace=True)
    print(f"removed duplicates from {df_name}")
    return df

remove_duplicates(players_df, "Players Info")
remove_duplicates(stats_df, "Players Stats")
```

Resetting the index so there is no irregularity like missing an index value.

```python
def reset_index(df, df_name):
    df = df.reset_index(drop=True)
    print(f"Index reset for {df_name}")
    return df
players_df = reset_index(players_df, "players_df")
stats_df = reset_index(stats_df, "stats_df")
```

Working on NULL_values, droping profil pic and players common name because most of it is null

```python
players_df['player_teams'] = players_df['player_teams'].fillna('Unknown')
print(players_df['player_teams'].isnull().sum())
players_df = players_df.drop(['profile_pic', 'player_common_names'], axis=1)
print(players_df.head())
```

found that some datetime data types were string converting them to datetime 

```python
date_column = ['born_date', 'debut_date', 'last_date']

for col in date_column:
    players_df[col] = pd.to_datetime(players_df[col], errors='coerce')

print(players_df.info())
```

Checking for impossible values  (outliers)

```python
print(players_df["height"].describe())
print(players_df["weight"].describe())
print(players_df["debut_age"].describe())
```

found an impossible value in player_weights, checking and fixing them.

```python
players_df.loc[players_df['weight'] < 50, 'weight'] = np.nan
print(players_df["weight"].describe())
print(players_df["weight"].isnull().sum())
players_df['weight'] = players_df['weight'].fillna(players_df['weight'].median())
```

removing spaces if any

```python
object_cols = players_df.select_dtypes(include="object").columns

for col in object_cols:
    players_df[col] = players_df[col].str.strip()
    
players_df.replace("", np.nan, inplace=True)

players_df.to_csv("cleaned_players_info.csv", index=False)
```

Saving Cleaned data of Players Info

```python
players_df.to_csv("cleaned_players_info.csv", index=False)

print(players_df.info())
```

Now onto the Players Stats Data set

Fixing player_id column for data type because some records have ID_ at start 

```python
stats_df['player_id'] = stats_df['player_id'].astype(str).str.replace('ID_', '').astype(int)
print(stats_df['player_id'].dtype)
```

Handling Missing values from Player stats

```python
stats_df["team"] = stats_df["team"].fillna("Unknown")
stats_df["team"] = stats_df["team"].str.strip()
```

Columns where missing likely means the stat wasn't recorded or the player had none

```python
zero_fill = [
    "goal_assists",
    "avg_goal_assists",
    "brownlow_votes",
    "hit_outs",
    "avg_hit_outs",
    "bounces",
    "avg_bounces",
    "marks_inside_50"
]

for col in zero_fill:
    stats_df[col] = stats_df[col].fillna(0)

print(stats_df.describe())
```

Percentage columns used median instead of 0 because values must be missing.

```python
median_fill = [
    "total_percentage_played",
    "avg_percentage_played"
]

for col in median_fill:
    if col in stats_df.columns:
        stats_df[col] = stats_df[col].fillna(stats_df[col].median())
```

Converting column into numeric if not any and filling with median because data might not be recorded

```python
cols = stats_df.columns[4:]

stats_df[cols] = stats_df[cols].apply(
    pd.to_numeric,
    errors="coerce"
)

numeric_cols = stats_df.select_dtypes(include="number").columns

for col in numeric_cols:
    stats_df[col] = stats_df[col].fillna(stats_df[col].median())

print(stats_df.isnull().sum())
```

Checking for outliers.

```python
print((stats_df["games_played"] < 0).sum())
stats_df = stats_df[stats_df["games_played"] >= 0]
```

Removing negative values or outliers

```python

numeric = stats_df.select_dtypes(include="number").columns

# Removed negative values
for col in numeric:
    stats_df = stats_df[stats_df[col] >= 0]
```

Checking duplicate player year records

```python
stats_df.duplicated(
    subset=["player_id","year","is_finals"]
).sum()
```

Verifying stats column after numeric conversion

```python
stats_df.describe()
```

Validation

```python
print("Rows:", len(stats_df), "\n")

print("Missing Values")

print(stats_df.isnull().sum(), "\n")

print("Duplicates")

print(stats_df.duplicated().sum(), "\n")

print("Data Types")

print(stats_df.dtypes, "\n")
```

```python
print("Missing Values in players_df:")
print(players_df.isnull().sum())
print("\nMissing Values in stats_df:")
print(stats_df.isnull().sum())
```

Merging cleaned_datasets

```python
players_df.rename(columns={"id": "player_id"}, inplace=True)
merged = pd.merge(
    players_df,
    stats_df,
    on="player_id",
    how="left"
)
```
