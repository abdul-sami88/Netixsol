```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('titanic_cleaned.csv')
df.head()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df['Survived'])
sns.set_style('dark')
plt.title('Distribution of Survived Column')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

This histogram shows the survival of passengers in a simple way, as it is a boolean column which suggests that survival is either a yes or no, and based on data most people didn't survive


```python
plt.figure(figsize=(6, 4))
sns.kdeplot(data=df['Survived'], fill=True)
sns.set_style('dark')
plt.title('Kernel Density Estimate of Survived Column')
plt.show()
```

this is just like the previous one but a KDE plot that shows data about passenger survival

```python
grouped_cat = df.groupby('ClassName')['Survived'].mean().reset_index()
sns.barplot(data=grouped_cat, x='ClassName', y='Survived')
sns.set_style('dark')
plt.title('Average Survival Rate by Class')
plt.xlabel('Class Name')
plt.ylabel('Average Survival Rate')
plt.show()
```

This bar chart shows the average survival rate with different passenger classes. It appears as the the 1st class has the most survival rate comparing to other lower classes.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True)
sns.set_style('dark')
plt.title('Correlation Heatmap of Numeric Columns')
plt.xlabel('Columns')
plt.ylabel('Columns')
plt.show()
```

The heatmap basically shows the corelation between variables of  numeric features. the darker and light shades shows which are closer to one another

```python
sns.boxplot(x=df["Age"])
sns.set_style('dark')
plt.xlabel("Age")
plt.title("Distribution of Passenger Ages")
plt.show()
```

This box plot shows us that we got quite some outliers in this column, most passengers ages lie in adult range of around 20 plus and less than 40. the whiskers strech from almost 3 to 55.
