```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

Q.1 How many Male and Female were there?
There were 577 males and 314 females.

```python
total_count = df["Sex"].value_counts()
print(total_count)
```

Q2. How many percentage of Passengers Survived?
Approximately 38.38% of passengers survived.

```python
ppl_sur = df["Survived"].value_counts(normalize=True)
print(ppl_sur)
```

Q3. How many males and how many females survived?
74.20% of females survived, and 18.89% of males survived.

```python
gender_sur = df.groupby('Sex')['Survived'].mean()
print(gender_sur)
```

Q4. What is survival rate by passenger class?
Survival rates by passenger class:
- Class 1: 62.96%
- Class 2: 47.28%
- Class 3: 24.24%

```python
class_survival = df.groupby("Pclass")["Survived"].mean()

print(class_survival)
```

Q5. How many were children and how many survived?
There were 113 children (under 18) on board, and 61 of them survived.

```python
children = df[(df["Age"] < 18)]

print("Total children :", len(children))

survived_children = children[children["Survived"] == 1]

print("Survived Children :", len(survived_children))

```

Q6. Who paid the highest fare?
The passengers who paid the highest fare are:
- Cardeza, Mr. Thomas Drake Martinez: 512.33
- Ward, Miss. Anna: 512.33
- Lesurer, Mr. Gustave J: 512.33

```python
highest_fare = (
    df[["Name", "Fare"]]
    .sort_values(by="Fare", ascending=False)
)

print(highest_fare.head(10))
```

Q.7 Where Passengers Embarked from and which had maximum passengers?
Passengers embarked from:
- Cherbourg (C): 168 passengers
- Queenstown (Q): 77 passengers
- Southampton (S): 644 passengers

The maximum passengers embarked from Southampton (S).

```python
embarked_class = df.groupby("Embarked")["Embarked"].count()
print("Embarked :", embarked_class)
max_embarked = embarked_class.idxmax()
print(max_embarked)
```

Q.8 Ratio of how many survived from each embark
Survival rates by embarkation point:
- Cherbourg (C): 55.36%
- Queenstown (Q): 38.96%
- Southampton (S): 33.70%

```python
embarked_survival = df.groupby('Embarked')['Survived'].mean()
print(embarked_survival)
```
